Para definir o coeficiente angular da regressão, faço:

a: coeficiente linear <br>
n: número de elementos<br>
Y = Moscow<br>
X = datas<br><br>

O coeficiente angular pode ser obtido usando a equação:<br><br>

$$ â =\frac{  n*Soma(Xi*Yi) - Soma(Xi)*Soma(Yi) } {n*Soma(Xi²) - (Soma(Xi))²} $$

(no caminho de aprender numpy, aprendendo LaTeX xD)

```python
Y = Moscow
X = datas
n = np.size(Moscow)    # aqui também posso usar Moscow.size, sem parênteses
```

```python
X**2  # eleva elemento a elemento de X ao quadrado.
```

```python
a = ( n*np.sum(X*Y) - np.sum(X)*np.sum(Y) ) / (n*np.sum(X**2) - np.sum(X)**2)  #conforme a fórmula acima
```

Coeficiente linear:

$ b = Media(Yi) - â*Media(Xi)   $

```python
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X + b          # agora sim foi a regressão linear.
```

```python
np.linalg.norm(Moscow-y)   # erro menor do que lá de cima. Suponho que seja o melhor erro possível (o que é estranho, eu esperaria erro 0)
```
