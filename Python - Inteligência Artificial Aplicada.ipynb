````python
import os
from google.colab import userdata
os.environ["GOOGLE_API_KEY"] = userdata.get('chave_teste')

from google import genai
client = genai.Client()

def resumir_email(lista):
  lista_de_resumos = []

  for i, email in enumerate(lista):
    resumo = client.models.generate_content(
        model="gemini-2.5-flash",
        contents= f"resuma em apenas uma linha o seguinte e-mail: {email}"
        )
    resposta = f"{i+1}. {resumo.text}"
    print(resposta)
    print("-" * 50)

    lista_de_resumos.append(resposta + "\n")
  return lista_de_resumos
````

````python
import pandas as pd
import numpy as np
import random

# Definindo o número de linhas
num_linhas = 50

# --- Gerando Dados Fictícios ---

# 1. Nome do Produto (Nomes aleatórios simples)
nomes_base = ['Notebook', 'Mouse Sem Fio', 'Teclado Mecânico', 'Monitor UltraWide', 'Webcam HD', 'Fone de Ouvido Bluetooth', 'SSD 1TB', 'Roteador Wi-Fi 6', 'Placa de Vídeo RTX', 'Cadeira Gamer']
nomes_produtos = [f'{random.choice(nomes_base)} {i+1}' for i in range(num_linhas)]

# 2. Categoria do Produto (Escolha aleatória de categorias)
categorias = ['Eletrônicos', 'Periféricos', 'Hardware', 'Acessórios']
categoria_produto = np.random.choice(categorias, size=num_linhas)

# 3. Preço do Produto (Números decimais aleatórios)
# Gerando preços entre 50.00 e 5000.00
preco_produto = np.round(np.random.uniform(50.0, 5000.0, size=num_linhas), 2)

# 4. Quantos itens do produto foram vendidos (Números inteiros aleatórios)
# Gerando vendas entre 10 e 1000
itens_vendidos = np.random.randint(10, 1001, size=num_linhas)

# 5. Avaliação do Produto (Números decimais aleatórios para avaliação)
# Gerando avaliações entre 1.0 e 5.0
avaliacao_produto = np.round(np.random.uniform(1.0, 5.0, size=num_linhas), 1)

# --- Criando o Dicionário de Dados ---
dados = {
    'Nome do Produto': nomes_produtos,
    'Categoria do Produto': categoria_produto,
    'Preço do Produto': preco_produto,
    'Itens Vendidos': itens_vendidos,
    'Avaliação do Produto': avaliacao_produto
}

# --- Criando o DataFrame ---
df_produtos = pd.DataFrame(dados)

# --- Exibindo o DataFrame e sua estrutura ---
print(df_produtos.head())
print('\n--- Informações do DataFrame ---')
print(df_produtos.info())
print(f'\nO DataFrame possui {df_produtos.shape[0]} linhas e {df_produtos.shape[1]} colunas.')
````

# Exceções -> try/except/finally

````python
# imagina que vamos tentar fazer algo errado, como converter para inteiro uma string "abc"

int("abc")

# sabemos que vai dar erro, nesse caso um "ValueError"
````

````python
# do mesmo jeito, no código abaixo, se eu digitar em vez de um número uma string, vai dar um "ValueError"

pegando_algo = int(input("Digite a sua idade: "))
````

````python
# existem vários outros tipos de erros. Por exemplo, se eu tentar somar string com ints

pessoas = 6
pessoas2 = "Mario, Joao, Silvia, Maria"

pessoas + pessoas2

# esse código vai dar um TypeError, um erro de tipo de variável
````

````python
# Para continuar o programa sem crashar existem recursos para tratar esses erros
# Por exemplo, no exemplo de pegando algo (ValueError):

try:
  pegando_algo = int(input("Digite a sua idade: "))
except:
  print("Você não digitou um número inteiro válido")

# inclusive é possível (e recomendável) criar as minhas próprias exceções
````

````python
# note que no exemplo acima eu não especifiquei o erro que era pra dar.
# Mais corretamente, meu código deveria enunciar exatamente o erro esperado

try:
  pegando_algo = int(input("Digite a sua idade: "))
  pessoas = 6
  pessoas2 = "Mario, Joao, Silvia, Maria"

  pessoas + pessoas2

except ValueError:
  print("Você não digitou um número inteiro válido")
except TypeError:
  print("Você não pode somar strings com ints")

# Nesse exemplo sempre vai dar TypeError, porque o erro está "no código" (e não no input). Só foi usado pra demonstrar como se encadeia exceções.
````

````python
# é possível dar "apelidos" aos erros usando "as"

try:
  pegando_algo = int(input("Digite a sua idade: "))
  pessoas = 6
  pessoas2 = "Mario, Joao, Silvia, Maria"

  pessoas + pessoas2

except ValueError as ve:  # ve como apelido de ValueError
  print("Você não digitou um número inteiro válido")
  print("Detalhes do erro:", ve) #ve se torna o texto do erro
except TypeError:
  print("Você não pode somar strings com ints")
````

````python
try:
  pegando_algo = int(input("Digite a sua idade: "))
  magica = 25/pegando_algo

  print(magica)

except ValueError as ve:
  print("Você não digitou um número inteiro válido")
  print("Detalhes do erro:", ve)
except TypeError:
  print("Você não pode somar strings com ints")

  # Nesse exemplo, existem exceções para ValueError e para TypeError, mas o código caiu num terceiro tipo de erro "não pensado"
  # para esses casos, é recomendado encontrar "Exception", que "pega todos os erros", caso não tenha caído em nenhum dos pensados
````

````python
try:
  pegando_algo = int(input("Digite a sua idade: "))
  magica = 25/pegando_algo

  print(magica)

except ValueError as ve:
  print("Você não digitou um número inteiro válido")
  print("Detalhes do erro:", ve)
except TypeError:
  print("Você não pode somar strings com ints")
except Exception as e: #Permite que o código não crashe, importante!
  print("DEU RUIM GERAL")
  print("Detalhes do erro:",e)
````

````python
# por fim, existe o finally, para fazer alguma coisa INDEPENDENTEMENTE de onde deu erro. (pra não perder o progresso no meio de um programa)
# finally roda SE DER ERRO OU NÃO

try:
  pegando_algo = int(input("Digite a sua idade: "))
  magica = 25/pegando_algo

  print(magica)

except ValueError as ve:
  print("Você não digitou um número inteiro válido")
  print("Detalhes do erro:", ve)
except TypeError:
  print("Você não pode somar strings com ints")
except Exception as e:
  print("DEU RUIM GERAL")
  print("Detalhes do erro:",e)
finally:          # é opcional. Geralmente tem try e except, mas finally só às vezes.
  print("Finalizei o processo")
````
